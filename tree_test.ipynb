```python
from sklearn.tree import DecisionTreeRegressor
import numpy as np
from matplotlib import pyplot as plt
from sklearn.tree import plot_tree

# Step 1: Create a synthetic dataset matching the specific rules
# Define sample points for each rule (considering the middle point for simplicity)
X_synthetic = np.array([
    [-1, -0],  # Rule 1 region
    [0.25, -0.5],   # Rule 2 region
    [-0.5, 0.5],   # Rule 3 region
    [0.5, 0.5]     # Rule 4 region
])

# Assign unique target values for each rule's region for simplicity
y_synthetic = np.array([1, 2, 3, 4])

# Step 2: Train the decision tree on the synthetic dataset
# Initialize the Decision Tree Regressor with a max_depth of 3 to match the rules
tree = DecisionTreeRegressor(max_depth=3, random_state=42)
tree.fit(X_synthetic, y_synthetic)

# Plot the tree to visualize if it matches the specified rules
plt.figure(figsize=(10, 5))
plot_tree(tree, filled=True, feature_names=['Feature 1', 'Feature 2'], proportion=False, rounded=True, precision=2, impurity=False)
plt.show()
```

```python
xx0_high_res, xx1_high_res = np.meshgrid(np.arange(-1,1, 0.01),  # Higher resolution
                                         np.arange(-1,1, 0.01))


# Predict over the high-resolution meshgrid with the depth-limited regressor
Z_depth_2 = tree.predict(np.c_[xx0_high_res.ravel(), xx1_high_res.ravel()])
Z_depth_2 = Z_depth_2.reshape(xx0_high_res.shape)

# Plot the results with depth limitation
plt.figure(figsize=(10, 6))
contour_depth_2 = plt.contourf(xx0_high_res, xx1_high_res, Z_depth_2, cmap='viridis', alpha=0.75)
plt.colorbar(contour_depth_2)
plt.title("Decision Tree Regressor Predictions with Max Depth 2")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
